```python
def afficher_doublons(df):
    """
    Détecte et affiche les lignes en double d'un DataFrame ainsi que leurs index.
    """

    pd.set_option('display.max_colwidth', None)

    # Trouver toutes les occurrences de doublons
    duplicates = df.duplicated(keep=False)

    # Extraire les lignes qui sont des doublons
    duplicated_rows = df[duplicates]

    if duplicated_rows.empty:
        print("Aucun doublon trouvé.")
    else:
        print("Index des lignes en double :")
        print(duplicated_rows.index.tolist())

        print("\nLignes en double trouvées :")
        print(duplicated_rows)
```

##III - Recherche d'irrégularités dans les colonnes et traitement

Code différent pour les variables catégorielle et continues (attention aux NA et Nan, ce code est adapté à la présence de Nan et non NA pour les variables numériques)

```python
# df reprennant les valeurs uniques de chaque variable catégorielle pour visualiser la présence d'irrégulatités
categorical_vars = coeur.select_dtypes(include=['category']).columns

# Préparation des données pour le DataFrame
data = [{"Variable": var, "Description": dict_categorielle[var], "Unique Values": coeur[var].unique()} for var in categorical_vars]

# Création du DataFrame en utilisant pandas.concat
irregularities_categorical_vars = pd.concat([pd.DataFrame([d]) for d in data], ignore_index=True)

# Configurer l'affichage pour montrer le contenu complet des cellules
pd.set_option('display.max_colwidth', None)
# Configurer l'affichage pour montrer le contenu complet des cellules, toutes les valeurs
pd.set_option('display.max_seq_items', None)

irregularities_categorical_vars['Unique Values'] = irregularities_categorical_vars['Unique Values'].apply(
    lambda x: np.concatenate((np.sort([y for y in x if pd.notna(y)]), [y for y in x if pd.isna(y)])))

# Affichage du DataFrame
irregularities_categorical_vars
```

Observations et modifications spécifiques à chaque dataset

**DETECTION** (valable pour les variables str)

```python
# Fonction pour détecter les irrégularités dans une cellule
def detect_irregularities(cell):
    if isinstance(cell, str):  # Vérifie si la cellule est une chaîne de caractères
        return any(c for c in cell if c in [' ', '\t', '?'] or not c.isprintable())
    return False

# Appliquer la fonction de détection à chaque cellule du DataFrame
irregularities = coeur.applymap(detect_irregularities)

# Trouver les colonnes et les lignes où il y a des irrégularités
irregular_cells = irregularities.any(axis=1)
irregular_columns = irregularities.any(axis=0)

# Afficher les résultats
print("Lignes avec irrégularités:", coeur[irregular_cells].index.tolist())
print("Colonnes avec irrégularités:", irregular_columns[irregular_columns].index.tolist())
```

```python
# df reprennant les valeurs uniques de chaque variable numériques pour visualiser la présence d'irrégulatités

# Sélection des colonnes de type int64 et float64, à l'exception de la colonne 'id'
continuous_vars = coeur.select_dtypes(include=['int64', 'float64']).columns
continuous_vars = [var for var in continuous_vars if var != 'id']

# Préparation des données pour le DataFrame
data = [{"Variable": var, "Description": dict_continues[var], "Unique Values": coeur[var].unique()} for var in continuous_vars]

# Création du DataFrame en utilisant pandas.concat
irregularities_continuous_vars = pd.concat([pd.DataFrame([d]) for d in data], ignore_index=True)

# Configurer l'affichage pour montrer le contenu complet des cellules
pd.set_option('display.max_colwidth', None)
# Configurer l'affichage pour montrer le contenu complet des cellules, toutes les valeurs
pd.set_option('display.max_seq_items', None)

# Tri de Unique Values en ordre ascendant, y compris NaN
irregularities_continuous_vars['Unique Values'] = irregularities_continuous_vars['Unique Values'].apply(
    lambda x: np.sort(x)  # Tri en incluant NaN
)

# Affichage du DataFrame
irregularities_continuous_vars
```

**TRAITEMENT** des blancs, tabulations, caractères spéciaux

```python
# Fonction pour nettoyer chaque cellule du df
def clean_cell(cell):
    # Si la cellule est vide (NaN), retourne telle quelle
    if pd.isnull(cell):
        return cell
    # Pour les colonnes numériques, garde seulement les chiffres et les points
    # Convertit les autres caractères en NaN
    if isinstance(cell, (int, float)):
        cleaned = ''.join(c for c in str(cell) if c.isdigit() or c == '.')
        # Convertit la chaîne nettoyée en float, ou en NaN si elle est vide
        return float(cleaned) if cleaned else float('nan')
    # Pour les colonnes textuelles, supprime tous les espaces et caractères spéciaux
    else:
        # élimine tous les caractères qui ne sont ni des lettres ni des chiffres
        return ''.join(c for c in str(cell) if c.isalnum())

# Applique la fonction de nettoyage à chaque cellule du DataFrame
coeur = coeur.applymap(clean_cell)

# Affiche les premières lignes du jeu de données nettoyé
coeur.head(20)
```

##V - Recherche des 0 pour les variables continues

Ici le code a dû être modifié par rapport aux autres datasets car les variables catégorielles, même modifiées en type category sont toujours considérées comme int ou float

```python
# Sélectionner toutes les colonnes numériques
colonnes_numeriques = coeur.select_dtypes(include=['int64', 'float64']).columns

# Exclure les colonnes qui ont été converties en type 'category'
colonnes_a_exclure = coeur.select_dtypes(include=['category']).columns
colonnes_avec_zeros = [col for col in colonnes_numeriques if col not in colonnes_a_exclure]

# Exclure les colonnes catégorielles
colonnes_avec_zeros = [col for col in colonnes_numeriques if col not in categorical_vars]

# Calcul du nombre et du pourcentage de zéros dans chaque colonne concernée
nb_zero = coeur[colonnes_avec_zeros].isin([0]).sum()
zero_percentage = (nb_zero / len(coeur)) * 100

# Création d'un DataFrame pour présenter les résultats avec le nom de la variable et sa description
df_zeros = pd.DataFrame({
    'Variable': colonnes_avec_zeros,
    'Description': [dict_continues[col] if col in dict_continues else col for col in colonnes_avec_zeros],
    'nb_zero': nb_zero,
    'Zero Percentage (%)': zero_percentage
}).sort_values(by='nb_zero', ascending=False)

df_zeros.set_index('Variable', inplace=True)

df_zeros

```

```python
from scipy.stats import chi2_contingency

# Liste des noms de colonnes catégorielles
colonnes_categorielles = {
    'sex': 'Genre',
    'cp': 'Douleur thoracique',
    'fbs': 'Glycémie à jeun élevée',
    'restecg': 'Résultats électrocardiographiques au repos',
    'exang': 'Angine provoquée par exercice',
    'slope': 'Pente du segment ST exercice',
    'ca': 'Nb vaisseaux sanguins colorés par la fluoroscopie',
    'thal': 'Type de thalassémie'
}

# Fonction pour effectuer le test du chi-carré et afficher les résultats
def chi_squared_test(coeur, colonne):
    contingency_table = pd.crosstab(coeur[colonne], coeur['target_str'])
    chi2, p, _, _ = chi2_contingency(contingency_table)

    print(f"Test du Chi-squared pour la variable '{colonne}':")
    print(f"Statistique du Chi2 = {chi2}")
    print(f"P-value = {p}")
    print("")

# Appliquez le test du chi-carré pour chaque colonne catégorielle
for colonne, label in colonnes_categorielles.items():
    chi_squared_test(coeur, colonne)
```
